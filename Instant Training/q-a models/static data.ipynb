```python
import time 
time.clock=time.time
from chatterbot import ChatBot
from chatterbot.trainers import ListTrainer
import pandas as pd
# Load your CSV data
data = pd.read_csv("static data.csv")
# Create a new ChatBot instance
chatbot = ChatBot('MyChatBot')
# Create a ChatterBotCorpusTrainer for the chatbot
trainer = ListTrainer(chatbot)
# Convert CSV data to ChatterBot corpus format and train the chatbot
for index, row in data.iterrows():
    trainer.train([row['Question'], row['Answer']])
print("Chatbot: Hi! I'm your chatbot. You can type 'exit' to end the conversation.")
while True:
    user_input = input("You: ")
    if user_input.lower() == 'exit':
        print("Chatbot: Goodbye!")
        break
    response = chatbot.get_response(user_input)
    print("Chatbot:", response)
```
