```python
import time
time.clock=time.time
from chatterbot import ChatBot
from chatterbot.trainers import ListTrainer
import pandas as pd
def expensive(user_input, data1):
    us_input = user_input.split()
    for i in us_input:
        if i == 'expensive':
            response = data1[['area', 'price']][data1['price'] == data1['price'].max()]
            return response
        if i == 'cheapest':
            response = data1[['area', 'price']][data1['price'] == data1['price'].min()]
            return response
data = pd.read_csv("static data.csv")
data1= pd.read_csv("homeprices.csv")
chatbot = ChatBot('MyChatBot')
trainer = ListTrainer(chatbot)
for index, row in data.iterrows():
    trainer.train([row['Question'], row['Answer']])
def get_response(input_text):
    response = chatbot.get_response(input_text)
    if response.confidence < 0.5:
        return "Sorry, I don't understand."
    return response.text
print("Chatbot: Hi! I'm your chatbot. You can type 'exit' to end the conversation.")
while True:
    user_input = input("You: ")
    if user_input.lower() == 'exit':
        print("Chatbot: Goodbye!")
        break
    response = expensive(user_input, data1)
    if response is not None:
        print("Chatbot:", response)
    else:
        response = get_response(user_input)
        print("Chatbot:", response)
```
